## Exploratory Data Analysis(EDA)

```python
#Plot for Product count
plt.figure(figsize=(10,6))
sns.countplot(y=df['Product'],order=df['Product'].value_counts().index,palette='viridis')
plt.title('Plot for Product Count')
plt.show()
```

Highest number of product sell is USB-C charging Cable

```python
# Count plot of City
plt.figure(figsize=(10,6))
sns.countplot(y=df['City'],order=df['City'].value_counts().index,palette='viridis')
plt.title('Count plot of City')
plt.show()
```

Highest sell  happen from Sanfransisco

```python
corr_matrix=df.select_dtypes(include=['number']).corr()
corr_matrix
```

```python
# correlation matrix
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='magma')
plt.title('correlation Matrix')
plt.show()
```

```python
# heatmpa for interaction between multiple features
#citywise hights product sales
pivot_table = df.pivot_table(values= 'Sales', index='Product', columns='City', aggfunc = 'sum')
plt.figure(figsize=(12,8))
sns.heatmap(pivot_table, annot=True,fmt='.0f',cmap='YlGnBu')
plt.title('Heatmap of Sales by Product and City')
plt.show()
```

In San Francisco Macbook Pro LaptopHight selling

```python
df['Sales']=df['Quantity Ordered'].astype('int')*df['Price Each'].astype('float')
```

```python
#add new columns name as months by extracting month from orderdate
df['Months']=pd.to_datetime(df['Order Date']).dt.month.astype("int32")
df.head(5)
```

```python
df['Time'] = df['Order Date'].dt.strftime('%H:%M:%S')
df.head(5)
```

```python
#df['Month']=pd.to_datetime(df['Order Date']).dt.month.astype('int')
```

```python
# Add hour column
df['Hours'] = pd.to_datetime(df['Order Date']).dt.hour
df['Minute'] = pd.to_datetime(df['Order Date']).dt.minute
df['Count'] = 1
df.head()
```

```python
df['Sales']=df['Quantity Ordered'].astype('int')*df['Price Each'].astype('float')
df = df.drop(columns=['Order Date']) 
```

```python
#df.groupby(['Months']).sum()
```

```python
months = range(1,13)
print(months)
plt.bar(months,df.groupby(['Months']).sum()['Sales'],color='Skyblue')
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month number')
plt.show()
```
